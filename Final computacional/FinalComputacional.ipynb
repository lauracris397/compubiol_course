#### Resumen


Algunos rasgos comportamentales pueden ser utilizados para describir la historia evolutiva de los diferentes organismos, como la telaraña para la reconstrucción filogenética del orden Araneae. En este caso se quiso evaluar el soporte utilizado por la familia Trochilidae como un rasgo que da cuenta de las relaciones filogenéticas dentro de la familia, para esto se construyó un árbol consenso al que se le agrego información del soporte utilizado por cada especie de colibrí y vencejo (grupo externo), observando si los diferentes linajes daban cuenta del uso de cierto tipo de soporte o no. En general, el soporte utilizado por esta familia puede ser un abrebocas para mostrar relaciones filogenéticas y dar cuenta de la historia evolutiva de esta familia, pero son necesario tener en cuenta una mayor cantidad de factores si lo que se pretende es reconstruirla desde cero. 


#### Introducción 

A pesar de que en la actualidad se considera que los caracteres morfológicos y moleculares son rasgos mucho más precisos para estimar relaciones evolutivas, que el uso de rasgos relacionados al comportamiento, gran cantidad de autores han reconstruido relaciones evolutivas entre taxa con derivados de rasgos comportamentales, un ejemplo de esto es el uso de las telas de araña (Blamires 2006, Kuntner et al. 2008) y los avisperos (Bailey 2009). No obstante, varios autores aseguran que el comportamiento no provee la suficiente información acerca de las relaciones evolutivas por ser bastante lábil (Urbani 1989, Lee et al. 1996, Blomberg et al. 2003). Es entonces donde estudios como el presente deben aferrarse a posiciones como las del biólogo evolutivo Richard Dawkins, quien en 1982 propone el concepto de Fenotipo Extendido (FE), en el cual argumenta que, así como hay genes demostrables para la forma corporal o el color de los ojos, debe haber genes cuya expresión fenotípica se hace aparente directamente en el comportamiento. 
Por otro lado, las aves son un grupo bastante estudiado con aproximadamente 10.000 especies conocidas (Flores et al. 2014), donde un país como Colombia puede hacer alarde de poseer su mayor diversidad. Los estudios en este grupo son bastante enfocados en documentar las especies con su respectiva dieta, sin embargo, hay poco acerca de su comportamiento como tal, el cual es importante para entender el uso de recursos, procesos de adaptación y requisitos básicos (Delagado- V & Correa-H 2013). Las aves son vistas como un grupo carismático, descrito frecuentemente con los adjetivos vistoso y hermoso; no solo las hacen llamativas sus colores y melódicos cantos, sino también las obras arquitectónicas magnificas que pueden llegar a construir, para autores como Flores et al.  (2014) sus nidos no son más que proezas de construcción, los cuales poseen gran cantidad de materiales, utilizados para dar soporte, forma, mantener la temperatura y camuflarse con su entorno; son también varias la ubicación de su construcción, que van desde la vegetación, como a cavidades de troncos muertos, barrancos e inclusive el suelo. 
En varias ocasiones, respaldándose en que los nidos son una parte del fenotipo extendido, diferentes autores han utilizado los nidos para reconstruir filogenias, donde se han obtenido resultados contrastantes, por ejemplo, Lee et al. en 1996 encontraron que este tipo de caracteres no son filogenéticamente informativos dentro de la tribu Colcaliini en la familia de los vencejos, resultado similar a lo obtenido por Zyskowski & Prum en 1999 para la familia Furnariidae, en donde observan que la variación en la construcción de los nidos se encontraba más nivel de género, que de especie. A pesar de esto, muchos estudios, como los realizados por Queiroz & Wimberger en 1993, Winkler & Sheldon en 1993 y Sturmbauer et al. en 1998, han demostrado que los nidos de las aves realmente son útiles como caracteres filogenéticos. 
Apoyándonos en la premisa de que los nidos poseen características especie-especificas (Collias & Collias 1984) y en lo propuesto por Richard Dawkins, es posible pensar que hay bases genéticas asociadas en la construcción de los nidos y son entonces un rasgo útil para la reconstrucción de filogenias. Por lo que en este trabajo se pretende realizar un análisis comparativo filogenético teniendo en cuenta algunos caracteres que describan la arquitectura de los nidos de la familia Trochilidae, especialmente el soporte que utilizan. 



#### Metodología

Se realizaron 2500 árboles filogenéticos comparativos por medio de la página web birdtree.org (Jetz et al. 2012) de 43 especies de colibrí y 4 especies de vencejos, estos últimos fueron usados como grupos externos; este número estuvo restringido a las especies de las cuales se encontró información acerca de su nido, la cual se tomó de los anexos de la tesis de grado de Tigreros-Andrade (2017) y la guía de nidos de las aves de Remedios, Antioquia (2014). Posteriormente se realizó el árbol consenso por medio TreeAnnotator de BEAST y mediante el uso del programa FigTree se pudo visualizar el árbol consenso y agregar el estadístico de probabilidad posterior. 
Obtenido el árbol se agregó en este la información obtenida de cada tipo de nido, la cual se diferencia en 4 tipos de soportes que eran utilizados por estas aves (Tigreros-Andrade 2017), los cuales son nidos soportados en estructura vertical similar a la pared, nidos sujetos a la parte apical de hojas de diferentes especies vegetales (soporte lateral superior), nidos con soporte inferior que incluye nidos soportados en su base tanto encima de hojas como en la rama de algunas plantas y nidos con soporte superior que incluye nidos con un tipo de anclaje en el que se sujeta solo una porción superior del mismo.

Sin embargo, para el resto de los linajes a excepción de Haplophaedia aureliae se obtuvo su máximo posible que es 1.

Estos resultados parecen ser congruentes con lo obtenido por Tigreros-Andrade en 2017, nuevamente se ve una relación en la evolución de las especies de los colibríes con el tipo de soporte utilizado para anclar sus nidos. Esto se debe a que se encontraron diferentes clados en los que se compartían estructuras similares de soporte, por ejemplo, en los ermitaños hay presencia única del soporte lateral superior. Lo cual puede deberse a que estos son un grupo que disminuye en abundancia y diversidad hacia altas elevaciones o áreas secas y su alta dominancia y diversidad se concentra en los bosques húmedos de tierras bajas (Stiles 1981, Arizmendi & Rodríguez-Flores 2012) indicando que posiblemente los eventos de convergencia de este tipo de soporte pudieron haberse originado con relación al tipo de hábitat de estas especies (Tigreros-Andrade 2017).
Se cree que el soporte inferior debe ser el tipo de soporte utilizado por el ancestro de todos los colibríes, ya que es menos complejo porque solo implica la estructuración de los materiales encima de una base determinada, por lo que son requeridas menos habilidades motoras en comparación con los otros nidos. Autores como Collias & Collias (1984) aseveran que la selección natural puede favorecer los nidos que tienden a economizar esfuerzos, por lo que nidos con soporte más simple pueden verse favorecidos, explicando así la amplia distribución de este tipo de nidos en la familia Trochilidae. 


#### Conclusión 

Los nidos de colibríes, en este caso el tipo de soporte utilizado, pueden ayudar a describir la historia evolutiva de cada una de los linajes de esta familia, deben ser tomados en cuenta mayor cantidad de rasgos para asegurar mayor precisión, pero son un buen abrebocas para relacionar ciertos linajes, ya algunos parecen mantener un mismo tipo de soporte sin variación dentro del grupo.  En futuras investigaciones además del tipo de soporte utilizado, se deben incluir características como forma, tamaño y quizás tipos de materiales utilizados, haciendo una respectiva comparación de cuál de estos puede ser un rasgo que cuente mejor la historia evolutiva de la familia Trochilidae.
